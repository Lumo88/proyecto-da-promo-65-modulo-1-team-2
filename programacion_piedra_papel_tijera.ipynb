```python
ronda = 0 # Variables iniciales cada vez que comienza la partida
puntos_jugador = 0
puntos_maquina = 0

import random # Activamos la funciÃ³n 'aleatoria' para que la mÃ¡quina pueda jugar
opciones_palabras = ["piedra", "papel", "tijera"] # Fijamos una lista con las Ãºnicas palabras necesrias para jugar

while ronda < 3: #  Condicional 1: Comenzamos con el juego, para saber si comienza la ronda o no
    print(f"Marcador --> Jugador: {puntos_jugador} / MÃ¡quina: {puntos_maquina}")


    ronda = ronda + 1
    print(f"RONDA {ronda}") # Hay <= 3 rondas, por lo tanto, la partida continÃºa 

    palabra_jugador = input("Elige 'Piedra', 'Papel' o 'Tijera' para jugar en esta ronda").lower() # El jugador elige una opciÃ³n de la lista existente
    print("La palabra que has elegido como jugador es:  ", palabra_jugador)
    palabra_maquina = random.choice(opciones_palabras) # La mÃ¡quina elige una opciÃ³n de manera aleatoria de la lista existente
    print("La palabra que ha elegido la mÃ¡quina es:  ", palabra_maquina)

    if palabra_jugador == palabra_maquina: # Condicional 2: Comparamos las palabras elegidas, para determinar al ganador de cada ronda de la partida
        print("Â¡Empate!") # Empate y se vuelve a iniciar la ronda

    elif (palabra_jugador == "piedra" and palabra_maquina == "tijera") or (palabra_jugador == "tijera" and palabra_maquina == "papel") or (palabra_jugador == "papel" and palabra_maquina == "piedra"):
        print("Â¡El jugador ha ganado!") # Para que el 'jugador' gane, se suma el punto del 'jugador' y se vuelve a iniciar la ronda
        puntos_jugador = puntos_jugador + 1

    else:
        print("Â¡La mÃ¡quina ha ganado!") # Para que la 'mÃ¡quina' gane, se suma el punto de la 'mÃ¡quina' y se vuelve a iniciar la ronda
        puntos_maquina = puntos_maquina + 1

else: # Hay > 3 rondas, por lo tanto, la partida finaliza y se determina al ganador de la partida
    if puntos_jugador > puntos_maquina: 
        print("El jugador ha ganado la partida ğŸ‰ğŸ¥¸")
    else:
        print("La mÃ¡quina ha ganado la partida ğŸ‰ğŸ¤–")
```
