PROGRAMACIÓN JUEGO "PIEDRA, PAPEL O TIJERA"

```python
ronda = 0 # Variables iniciales cada vez que comienza la partida
puntos_jugador = 0
puntos_maquina = 0

import random # Activamos la función 'aleatoria' para que la máquina pueda jugar
opciones_palabras = ["piedra", "papel", "tijera"] # Fijamos una lista con las únicas palabras necesrias para jugar

while ronda < 3: #  Condicional 1: Comenzamos con el juego, para saber si comienza la ronda o no

    ronda = ronda + 1 # Hay <= 3 rondas, por lo tanto, la partida continúa 

    palabra_jugador = input("Elige 'Piedra', 'Papel' o 'Tijera' para jugar en esta ronda").lower() # El jugador elige una opción de la lista existente
    print("La palabra que has elegido como jugador es:  ", palabra_jugador)
    palabra_maquina = random.choice(opciones_palabras) # La máquina elige una opción de manera aleatoria de la lista existente
    print("La palabra que ha elegido la máquina es:  ", palabra_maquina)

    if palabra_jugador == palabra_maquina: # Condicional 2: Comparamos las palabras elegidas, para determinar al ganador de cada ronda de la partida
        print("¡Empate!") # Empate y se vuelve a iniciar la ronda

    elif (palabra_jugador == "piedra" and palabra_maquina == "tijera") or (palabra_jugador == "tijera" and palabra_maquina == "papel") or (palabra_jugador == "papel" and palabra_maquina == "piedra"):
        print("¡El jugador ha ganado!") # Para que el 'jugador' gane, se suma el punto del 'jugador' y se vuelve a iniciar la ronda
        puntos_jugador = puntos_jugador + 1

    else:
        print("¡La máquina ha ganado!") # Para que la 'máquina' gane, se suma el punto de la 'máquina' y se vuelve a iniciar la ronda
        puntos_maquina = puntos_maquina + 1

else: # Hay > 3 rondas, por lo tanto, la partida finaliza y se determina al ganador de la partida
    if puntos_jugador > puntos_maquina: 
        print("El jugador ha ganado la partida")
    else:
        print("La máquina ha ganado la partida")
```
