```python
def jugar_piedra_papel_tijera():

    ronda = 0 # Variables iniciales cada vez que comienza la partida
    puntos_jugador = 0
    puntos_maquina = 0

    import random # Activamos la funci√≥n 'aleatoria' para que la m√°quina pueda jugar
    opciones_palabras = ["piedra", "papel", "tijera"] # Fijamos una lista con las √∫nicas palabras necesrias para jugar

    print() # DAR LA BIENVENIDA AL JUGADOR AL JUEGO --> EJ: BIENVENID@ AL JUEGO CL√ÅSICO DE PIEDRA, APPEL O TIJERA


    while ronda < 3: # Bucle prinicpal del juego --> Comenzamos con el juego, para saber si comienza la ronda o no --> Hay < 3 rondas, por lo tanto, la partida contin√∫a 

        palabra_jugador = input("Elige 'Piedra', 'Papel' o 'Tijera' para jugar en esta ronda").lower() # El jugador elige una opci√≥n de la lista existente

        if palabra_jugador not in opciones_palabras: # En el caso de que el jugador no escriba correctamente alguna palabra de la lista, tendr√° que volverlo a escribir hasta que la escriba bien
            print(f"La palabra que has elegido: {palabra_jugador} no es v√°lida, vuelve a elegir entre 'Piedra', 'Papel' o 'Tijera")
            continue
        
        ronda += 1 # Contador de rondas
       
        if ronda == 3: # T√≠tulo de las rondas
            print("RONDA FINAL")
        else:
            print(f"RONDA {ronda}") 

        print(f"La palabra que has elegido como jugador es: {palabra_jugador}")
               
        palabra_maquina = random.choice(opciones_palabras) # La m√°quina elige una opci√≥n de manera aleatoria de la lista existente
        print(f"La palabra que ha elegido la m√°quina es: {palabra_maquina}")

        if palabra_jugador == palabra_maquina: # Condicional 1: Comparamos las palabras elegidas de cada jugador, para determinar al ganador de cada ronda de la partida
            print("¬°Empate!") # Empate y se vuelve a iniciar la ronda

        elif (palabra_jugador == "piedra" and palabra_maquina == "tijera") or (palabra_jugador == "tijera" and palabra_maquina == "papel") or (palabra_jugador == "papel" and palabra_maquina == "piedra"):
            print("¬°El jugador ha ganado!") # El 'jugador' gana, se suma el punto del 'jugador' y se vuelve a iniciar la ronda
            puntos_jugador += 1

        else:
            print("¬°La m√°quina ha ganado!") # La 'm√°quina' gana, se suma el punto de la 'm√°quina' y se vuelve a iniciar la ronda
            puntos_maquina += 1

        print(f"Marcador --> Jugador: {puntos_jugador} / M√°quina: {puntos_maquina}") # Mostramos los puntos de cada jugador por cada ronda

    if puntos_jugador > puntos_maquina: # Hay = 3 rondas, por lo tanto, la partida finaliza y se determina al ganador de la partida
        print("El jugador ha ganado la partida üéâü•∏")

    elif puntos_jugador < puntos_maquina:
        print("La m√°quina ha ganado la partida üéâü§ñ")

    else:
        print("Es un empate")

    print() # MENSAJE FINAL DE DESPEDIDA --> EJ: ESPERAMOS QUE VUELVA A JUGAR OTRA PARTIDA CON NOSOTROS
```
