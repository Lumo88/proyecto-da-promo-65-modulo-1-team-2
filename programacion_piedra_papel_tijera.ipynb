```python
ronda = 0 # Variables iniciales cada vez que comienza la partida
puntos_jugador = 0
puntos_maquina = 0

import random # Activamos la funciÃ³n 'aleatoria' para que la mÃ¡quina pueda jugar

##Imports para la visualizaciÃ³n de la interfaz de usuario.
import tkinter as tk
from tkinter import simpledialog
import ast
 
#variables inicializadas para la interfaz.
root = tk.Tk()
root.withdraw()
 
opciones_palabras = ["piedra", "papel", "tijera"] # Fijamos una lista con las Ãºnicas palabras necesrias para jugar

print( "ğŸ®ğŸª¨ğŸ“„âœ‚ï¸ BIENVENID@ AL CLÃSICO JUEGO DE PIEDRA, PAPEL O TIJERA ğŸ®ğŸª¨ğŸ“„âœ‚ï¸") 

print("ğŸ”¥PrepÃ¡rate para empezar, tendrÃ¡s 3 rondasğŸ”¥")


while ronda < 3: # Bucle princicpal del juego --> Comenzamos con el juego, para saber si comienza la ronda o no --> Hay < 3 rondas, por lo tanto, la partida continÃºa 

    # Usamos try para recoger la excepciÃ³n si el usuario cierra ventana o presiona Cancelar. Cambiamos el input() por simpledialog.askstring().
    try:
        palabra_jugador = simpledialog.askstring("Juego Piedra, Papel y Tijera", "Elige 'PiedrağŸª¨', 'PapelğŸ“„' o 'Tijeraâœ‚ï¸' para jugar en esta ronda").lower() # El jugador elige una opciÃ³n de la lista existente
    except:
        break

    if palabra_jugador not in opciones_palabras: # En el caso de que el jugador no escriba correctamente alguna palabra de la lista, tendrÃ¡ que volverlo a escribir hasta que la escriba bien
        print("--------------------------------------------------------------------------------")
        print(f"La palabra '{palabra_jugador}' no es vÃ¡lida")
        continue
        
    ronda += 1 # Contador de rondas
       
    if ronda == 3: # TÃ­tulo de las rondas
        print("--------------------------------------------------------------------------------")
        print("RONDA FINAL")

    else:
        print("--------------------------------------------------------------------------------")
        print(f"RONDA {ronda}")
           
    print(f"Has escogido: {palabra_jugador}")
               
    palabra_maquina = random.choice(opciones_palabras) # La mÃ¡quina elige una opciÃ³n de manera aleatoria de la lista existente
    print(f"La mÃ¡quina ha escogido: {palabra_maquina}")

    if palabra_jugador == palabra_maquina: # Condicional 1: Comparamos las palabras elegidas de cada jugador, para determinar al ganador de cada ronda de la partida
        print("Â¡EMPATE!ğŸ¤") # Empate y se vuelve a iniciar la ronda

    elif (palabra_jugador == "piedra" and palabra_maquina == "tijera") or (palabra_jugador == "tijera" and palabra_maquina == "papel") or (palabra_jugador == "papel" and palabra_maquina == "piedra"):
        print("Â¡El jugador ha ganado!ğŸ‘¤") # El 'jugador' gana, se suma el punto del 'jugador' y se vuelve a iniciar la ronda
        puntos_jugador += 1

    else:
        print("Â¡La mÃ¡quina ha ganado!ğŸ¤–") # La 'mÃ¡quina' gana, se suma el punto de la 'mÃ¡quina' y se vuelve a iniciar la ronda
        puntos_maquina += 1
        
    print(f"Marcador --> JugadorğŸ‘¤: {puntos_jugador} / MÃ¡quinağŸ¤–: {puntos_maquina}") # Mostramos los puntos de cada jugador por cada ronda

print("--------------------------------------------------------------------------------")

if puntos_jugador > puntos_maquina: # Hay = 3 rondas, por lo tanto, la partida finaliza y se determina al ganador de la partida
    print("El jugador ha ganado la partida ğŸ‰ğŸ‘¤")

elif puntos_jugador < puntos_maquina:
    print("La mÃ¡quina ha ganado la partida ğŸ‰ğŸ¤–")

else:
    print("Es un empateğŸ¤")

print("--------------------------------------------------------------------------------")
print("ğŸ‰Â¡Gracias por jugar a Piedra ğŸª¨ Papel ğŸ“„ o Tijera âœ‚ï¸! Esperamos verte pronto de nuevo ğŸ®ğŸ˜„") 
```
