```python
import random 

lista_palabras = [
    "python", "computadora", "programacion", "desarrollo", "algoritmo",
    "teclado", "pantalla", "software", "hardware", "internet",
    "servidor", "archivo", "variable", "funcion", "proyecto",
    "ahorcado", "adivinar", "palabra", "letras", "intentos",
    "victoria", "derrota", "consola", "terminal", "codigo",
    "libreria", "modulo", "interfaz", "usuario", "ejecutar"
] # le damos a python este listado de palabras, para que elija una al azar, que será la que adivine el usuario

palabra_secreta = random.choice(lista_palabras) # guardamos la palabra secreta

print("La palabra secreta---->",palabra_secreta) # print de control

intentos = 7 # número de intentos máximos del usuario para adivinar la palabra
fallos = 0 # contador de fallos, con el objetivo de comparar los fallos con los intentos, que será la condición de parada del while intentos == fallos

letras_introducidas = [] #aqui guardaremos todas las letras que el usuario vaya introduciendo
letras_acertadas = [] # aquí guardaremos las letras que el usuario vaya acertando, lo mostraremos en el tablero
letras_falladas = [] # aquí guardaremos las letras que el usuario vaya fallando, lo mostraremos en el tablero
palabras_falladas = [] # aquí guardaremos las palabras que el usuario vaya fallando, lo mostraremos en el tablero
tablero = "" # este será el tablero del juego, donde mostraremos el status del mismo tras cada iteración (o intento)

for a in palabra_secreta:
    tablero = tablero + "_" 

print("El estado actual del tablero es", tablero) # nuestro tablero inicial, mostrará _ por cada letra de la palabra

## letra_usuario = input("¡Introduce una letra (a-z) para salvar al Ahorcado!").lower() # le pedimos al usuario que introduzca una letra y la convertimos en minúscula



letra_usuario = "a"
print("Letra usuario ---> ",letra_usuario) # print de control
```

```python
if letra_usuario.isalpha() and len(letra_usuario) == 1: #comprobamos que lo introducido es letra
    print("¿Es letra?",letra_usuario.isalpha())

    if letra_usuario in letras_introducidas: #COMPROBAMOS SI YA LA HA INTRODUCIDO
        print("Ya has introducido esa letra, prueba una nueva") ## CAMBIAR POR UN INPUT
        print(f"Letras introducidas hasta el momento: {letras_introducidas}")

       
    else:
        letras_introducidas.append(letra_usuario)
        print("**Control**",letras_introducidas)


        if letra_usuario in palabra_secreta:
            print("estas entrando aqui")
            letras_acertadas.append(letra_usuario) #añadimos la letra en la lista de letras acertadas.
            
            pos = palabra_secreta.find(letra_usuario) #buscamos la posición en la palabra

            print ("pos previo while",pos)
            print ("longitud cadena palabra secreta",len(palabra_secreta))

            while pos != -1:
                print ("pos while",pos)
                print ("pos --->",pos,letra_usuario)
                tablero = tablero[:pos] + letra_usuario + tablero[pos+1:] #como es cadena, reescribimos con la letra encontrada.
                pos = palabra_secreta.find(letra_usuario, pos + 1)
            print("BIEEEEEEEEEN Has acertado la letra.")
            print("Intentos acertados: ", len(letras_acertadas), "de ", len(palabra_secreta))

        else:  ##EL USUARIO HA INTRODUCIDO UNA LETRA QUE NO ESTÁ

            letras_introducidas.append(letra_usuario) #añadimos la letra en lista total
            print("**Control**",letras_introducidas)
            letras_falladas.append(letra_usuario) #añadimos la letra en falladas
            intentos-=1 #restamos intento

            print("NOOOOOOOOOOO, no está esa letra en la palabra")
            print(f"Te quedan {intentos} intentos disponibles")
        
        print("TABLERO JUEGO: ",tablero)
           
else:
    print("El valor introducido no es una letra. Introduce una letra de a-z") # nos aseguramos de que el usuario haya introducido una letra y, si no lo es, devolvemos error

           
```
