```python
import random 

lista_palabras = [
    "python", "computadora", "programacion", "desarrollo", "algoritmo",
    "teclado", "pantalla", "software", "hardware", "internet",
    "servidor", "archivo", "variable", "funcion", "proyecto",
    "ahorcado", "adivinar", "palabra", "letras", "intentos",
    "victoria", "derrota", "consola", "terminal", "codigo",
    "libreria", "modulo", "interfaz", "usuario", "ejecutar"
] # le damos a python este listado de palabras, para que elija una al azar, que ser√° la que adivine el usuario

palabra_secreta = random.choice(lista_palabras) # guardamos la palabra secreta

print(" ‚úèÔ∏èLa palabra secreta---->",palabra_secreta) # print de control

intentos = 7 # n√∫mero de intentos m√°ximos del usuario para adivinar la palabra
fallos = 0 # contador de fallos, con el objetivo de comparar los fallos con los intentos, que ser√° la condici√≥n de parada del while intentos == fallos

letras_introducidas = [] #aqui guardaremos todas las letras que el usuario vaya introduciendo
letras_acertadas = [] # aqu√≠ guardaremos las letras que el usuario vaya acertando, lo mostraremos en el tablero
letras_falladas = [] # aqu√≠ guardaremos las letras que el usuario vaya fallando, lo mostraremos en el tablero
palabras_falladas = [] # aqu√≠ guardaremos las palabras que el usuario vaya fallando, lo mostraremos en el tablero
tablero = "" # este ser√° el tablero del juego, donde mostraremos el status del mismo tras cada iteraci√≥n (o intento)

for a in palabra_secreta:
    tablero = tablero + "_" 

print("El estado actual del tablero es", tablero) # nuestro tablero inicial, mostrar√° _ por cada letra de la palabra

## letra_usuario = input("¬°Introduce una letra (a-z) para salvar al Ahorcado!").lower() # le pedimos al usuario que introduzca una letra y la convertimos en min√∫scula



letra_usuario = "a"
print("Letra usuario ---> ",letra_usuario) # print de control

```

```python
if letra_usuario.isalpha() and len(letra_usuario) == 1: #comprobamos que lo introducido es letra
    print("‚åõ Procesando letra",letra_usuario.isalpha()) #print de control

    if letra_usuario in letras_introducidas: #COMPROBAMOS SI YA LA HA INTRODUCIDO
        print("Ya has introducido esa letra, prueba una nueva") ## CAMBIAR POR UN INPUT
        print(f"Letras introducidas hasta el momento: {letras_introducidas}")

       
    else:
        letras_introducidas.append(letra_usuario)
        print("üìùLa lista de letras introducidas es:",letras_introducidas)


        if letra_usuario in palabra_secreta:
            print("estas entrando aqui") #control
            letras_acertadas.append(letra_usuario) #a√±adimos la letra en la lista de letras acertadas.
            
            pos = palabra_secreta.find(letra_usuario) #buscamos la posici√≥n en la palabra

            print ("pos previo while",pos)
            print ("longitud cadena palabra secreta",len(palabra_secreta)) #se puede cambiar por el regex

            while pos != -1:
                print ("pos while",pos)
                print ("pos --->",pos,letra_usuario)
                tablero = tablero[:pos] + letra_usuario + tablero[pos+1:] #como es cadena, reescribimos con la letra encontrada.
                pos = palabra_secreta.find(letra_usuario, pos + 1)
            print("BIEEEEEEEEEN Has acertado la letra.")
            print("Intentos acertados: ", len(letras_acertadas), "de ", len(palabra_secreta))

        else:  ##EL USUARIO HA INTRODUCIDO UNA LETRA QUE NO EST√Å

            letras_introducidas.append(letra_usuario) #a√±adimos la letra en lista total
            print("**Control**",letras_introducidas)
            letras_falladas.append(letra_usuario) #a√±adimos la letra en falladas
            intentos-=1 #restamos intento

            print("NOOOOOOOOOOO, no est√° esa letra en la palabra")
            print(f"Te quedan {intentos} intentos disponibles")
        
        print("TABLERO JUEGO: ",tablero)
           
else:
    print("El valor introducido no es una letra. Introduce una letra de a-z") # nos aseguramos de que el usuario haya introducido una letra y, si no lo es, devolvemos error

           
```
