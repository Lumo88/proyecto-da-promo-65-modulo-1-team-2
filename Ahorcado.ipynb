```python
import random 
import warnings
warnings.filterwarnings("ignore", category=SyntaxWarning)

lista_palabras = [
    "python", "computadora", "programacion", "desarrollo", "algoritmo",
    "teclado", "pantalla", "software", "hardware", "internet",
    "servidor", "archivo", "variable", "funcion", "proyecto",
    "ahorcado", "adivinar", "palabra", "letras", "intentos",
    "victoria", "derrota", "consola", "terminal", "codigo",
    "libreria", "modulo", "interfaz", "usuario", "ejecutar"
] # le damos a python este listado de palabras, para que elija una al azar, que serÃ¡ la que adivine el usuario

palabra_secreta = random.choice(lista_palabras) # guardamos la palabra secreta

print(f"âœï¸La palabra secreta----> {palabra_secreta}") # demo

intentos = 7 # nÃºmero de intentos mÃ¡ximos del usuario para adivinar la palabra
fallos = 0 # contador de fallos, con el objetivo de comparar los fallos con los intentos, que serÃ¡ la condiciÃ³n de parada del while intentos == fallos

letras_introducidas = [] #aqui guardaremos todas las letras que el usuario vaya introduciendo
letras_acertadas = [] # aquÃ­ guardaremos las letras que el usuario vaya acertando, lo mostraremos en el tablero
lista_fallos = [] # aquÃ­ guardaremos las letras que el usuario vaya fallando, lo mostraremos en el tablero
tablero = "" # este serÃ¡ el tablero del juego, donde mostraremos el status del mismo tras cada iteraciÃ³n (o intento)

for a in palabra_secreta:
    tablero = tablero + "_" 
```

```python
def pintar_ahorcado(fallos):
   muneco_ahorcado = ['''
   +---+
       |
       |
       |
       ===''', '''
    +---+
    O   |
        |
        |
       ===''', '''
    +---+
    O   |
    |   |
        |
       ===''', '''
    +---+
    O   |
   /|   |
        |
       ===''', '''
    +---+
    O   |
   /|\  |
        |
       ===''', '''
    +---+
    O   |
   /|\  |
   /    |
       ===''', '''
    +---+
    O   |
   /|\  |
   / \  |
       ===''']
    
   print(muneco_ahorcado[fallos])
   
```

```python
print(f"El estado actual del tablero es {" ".join(tablero)}") # nuestro tablero inicial, mostrarÃ¡ " _" por cada letra de la palabra

while intentos > 0 and "_" in tablero:

    print("\n**************************\n")

    letra_usuario = input("Â¡Introduce una letra (a-z) o palabra completa para salvar al Ahorcado!").lower() # le pedimos al usuario que introduzca una letra y la convertimos en minÃºscula

    if letra_usuario.isalpha() and len(letra_usuario) == 1: #comprobamos que lo introducido es letra

        if letra_usuario in letras_introducidas: #COMPROBAMOS SI YA LA HA INTRODUCIDO
            print("Ya has introducido esa letra, prueba una nueva") ## CAMBIAR POR UN INPUT

        
        else:
            letras_introducidas.append(letra_usuario)

            if letra_usuario in palabra_secreta:
                letras_acertadas.append(letra_usuario) #aÃ±adimos la letra en la lista de letras acertadas.
                
                pos = palabra_secreta.find(letra_usuario) #buscamos la posiciÃ³n en la palabra

                while pos != -1:

                    tablero = tablero[:pos] + letra_usuario + tablero[pos + 1:] #como es cadena, reescribimos con la letra encontrada.
                    pos = palabra_secreta.find(letra_usuario, pos + 1)
                print("ğŸ‰BIEEEEEEEEENğŸ‰ Has acertado la letra.")
                print(f"TABLERO JUEGO: {" ".join(tablero)}") # Usamos .joint para mostrar el tablero de esta forma "_ _ _"

                if "_" not in tablero:
                    print(f"ğŸ¥‡Â¡ENHORABUENA!ğŸ¥‡ Has ganado. La palabra era: {palabra_secreta}")
                    break 

            else:  ##EL USUARIO HA INTRODUCIDO UNA LETRA QUE NO ESTÃ

                lista_fallos.append(letra_usuario) #aÃ±adimos la letra en falladas
                intentos -= 1 #restamos intento
                pintar_ahorcado(fallos)
                fallos += 1

                print("ğŸ˜¿NOOOOOOOOOOOğŸ˜¿ No estÃ¡ esa letra en la palabra")

                if intentos == 0:
                    print(f"â˜ ï¸NOOOOOOOOOOOâ˜ ï¸ GAME OVER - Has perdido (menos mal que es un juego ğŸ˜Œ). La palabra era {palabra_secreta}")
                    break 

                print(f"ğŸ¯Te quedan {intentos} intentosğŸ¯")
            
                print(f"TABLERO JUEGO: {" ".join(tablero)}") # Usamos .joint para mostrar el tablero de esta forma "_ _ _"
    
        print(f"ğŸ«£Fallos hasta el momento: {lista_fallos}")

    elif letra_usuario.isalpha() and len(letra_usuario) == len(palabra_secreta):
        if letra_usuario == palabra_secreta:
            tablero = palabra_secreta
            print(f"ğŸ¥‡Â¡ENHORABUENA!ğŸ¥‡ Has adivinado la palabra: {palabra_secreta}")
            break 
        
        else:
            lista_fallos.append(letra_usuario) #aÃ±adimos la letra en falladas
            intentos -= 1 #restamos intento
            pintar_ahorcado(fallos)
            fallos += 1

            print("ğŸ˜¿NOOOOOOOOOOOğŸ˜¿ Esta no es la palabra secreta")

            if intentos == 0:
                print(f"â˜ ï¸GAME OVERâ˜ ï¸ - Has perdido (menos mal que es un juego ğŸ˜Œ). La palabra era {palabra_secreta}.")
                break 

            print(f"ğŸ¯Te quedan {intentos} intentosğŸ¯")

        print(f"ğŸ«£Fallos hasta el momento: {lista_fallos}")

    else:
        print("âŒErrorâŒ Introduce un valor de a-z o una palabra con la longitud de la palabra secreta.") # nos aseguramos de que el usuario haya introducido una letra y, si no lo es, devolvemos error
```
