```python
import random 

#Import para que no aparezca el warning de un print.
import warnings
warnings.filterwarnings("ignore", category=SyntaxWarning)

##Imports para la visualizaciÃ³n de la interfaz de usuario.
import tkinter as tk 
from tkinter import simpledialog 
import ast

#variables inicializadas para la interfaz.
root = tk.Tk() 
root.withdraw()

# le damos a python este listado de palabras, para que elija una al azar, que serÃ¡ la que adivine el usuario
lista_palabras = [
    "python", "computadora", "programacion", "desarrollo", "algoritmo",
    "teclado", "pantalla", "software", "hardware", "internet",
    "servidor", "archivo", "variable", "funcion", "proyecto",
    "ahorcado", "adivinar", "palabra", "letras", "intentos",
    "victoria", "derrota", "consola", "terminal", "codigo",
    "libreria", "modulo", "interfaz", "usuario", "ejecutar"
]

palabra_secreta = random.choice(lista_palabras) # guardamos la palabra secreta

print(f"âœï¸La palabra secreta----> {palabra_secreta}") # demo

intentos = 7 # nÃºmero de intentos mÃ¡ximos del usuario para adivinar la palabra
fallos = 0 # contador de fallos para printear el muÃ±eco

letras_introducidas = [] #aqui guardaremos todas las letras que el usuario vaya introduciendo
letras_acertadas = [] # aquÃ­ guardaremos las letras que el usuario vaya acertando
lista_fallos = [] # aquÃ­ guardaremos las letras que el usuario vaya fallando
tablero = "" # este serÃ¡ el tablero del juego


for a in palabra_secreta: #Aqui inicializamos el tablero, tantos _ como letras de palabra.
    tablero = tablero + "_" 
```

```python
def pintar_ahorcado(fallos):
   ''' FunciÃ³n que recoge como parÃ¡metro fallos, para imprimir la posiciÃ³n de nuestro lista de estado del ahorcado'''
   muneco_ahorcado = ['''
   +---+
       |
       |
       |
       ===''', '''
    +---+
    O   |
        |
        |
        ===''', '''
    +---+
    O   |
    |   |
        |
        ===''', '''
    +---+
    O   |
   /|   |
        |
        ===''', '''
    +---+
    O   |
   /|\  |
        |
        ===''', '''
    +---+
    O   |
   /|\  |
   /    |
        ===''', '''
    +---+
    O   |
   /|\  |
   / \  |
        ===''']
    
   print(muneco_ahorcado[fallos])
   
```
